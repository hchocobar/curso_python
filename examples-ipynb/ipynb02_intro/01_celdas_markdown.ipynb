# Celdas Markdown

Markdown es un lenguaje de marcado ligero diseñado para que el texto sea legible y fácil de editar, incluso en su formato crudo.

Su objetivo principal es permitir la creación de contenido estructurado utilizando una sintaxis de texto plano que puede convertirse de manera sencilla a HTML o a otros formatos de visualización profesional. Al prescindir de herramientas complejas de edición visual, Markdown fomenta un flujo de trabajo ágil y enfocado en el contenido, convirtiéndose en el estándar de facto para la documentación técnica y la comunicación en entornos como Jupyter Notebook o plataformas de control de versiones.

Para implementar este formato, la sintaxis utiliza caracteres especiales que el intérprete traduce automáticamente.

## Jerarquía de Títulos

Se definen mediante el símbolo numeral (#) al inicio de la línea. A mayor cantidad de símbolos, menor es el nivel del título (hasta 6).

```markdown
# Título Principal (H1)
## Sección (H2)
### Subsección (H3)
```

# Título Principal (H1)
## Sección (H2)
### Subsección (H3)

## Énfasis y Estilo

Para resaltar información, utilizamos asteriscos o guiones bajos para definir el peso visual del texto.

```markdown
*Texto en cursiva*
**Texto en negrita**
***Texto en negrita y cursiva***
```

*Texto en cursiva*

**Texto en negrita**

***Texto en negrita y cursiva***

## Listas y Estructuración

Las listas permiten organizar conceptos técnicos de forma jerárquica o secuencial, facilitando la lectura rápida.

```markdown
* Elemento de lista no ordenada
* Otro elemento

1. Primer paso del proceso
2. Segundo paso del proceso
```

* Elemento de lista no ordenada
* Otro elemento

1. Primer paso del proceso
2. Segundo paso del proceso
