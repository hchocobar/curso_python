# Variables y Tipos de Datos

- Cadenas - str
- Enteros - int
- Flotantes (decimales) - float
- Booleanos - bool
- None - NoneType

## Cadena de texto (string)

```python
mi_cadena = "Hola Mundo!"

mi_cadena, type(mi_cadena)
```

```python
mi_cadena_larga = """
Esta es una cadena
de varias lineas """

mi_cadena_larga, type(mi_cadena_larga)
```

## Números enteros


```python
edad = 35

edad, type(edad)
```

## Número real o de punto flotante


```python
precio = 7435.99
porcentaje = .4

precio, porcentaje, type(porcentaje)
```

## Booleanos


```python
verdadero = True
falso = False

verdadero, falso, type(falso)
```

## None – NonType – Ausencia de un valor


```python
variable_sin_nada = None

variable_sin_nada, type(variable_sin_nada)
```
