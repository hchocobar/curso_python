Los paréntesis () modifican el orden de las operaciones


```python
print(14 + 15 * 3 - 16 / 4)
print(14 + 15 * (3 - 16) / 4)
print((14 + 15) * (3 - 16) / 4)
print((14 + 15) * 3 - 16 / 4)
```

Suma de enteros tiene com resultado un entero
Suma de enteros + decimal tiene como resultado un decimal
Suma de decimales tiene como resultado un decimal
