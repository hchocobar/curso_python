Para introducir un número decimal utilizamos la función integrada float()


```python
numero_decimal = float(input('Introduce un numero decimal: '))
print('El numero decimal que ingreso es:', numero_decimal)
```

Para convertir un número en cadena utilizamos la función integrada str()


```python
nueva_cadena = str(numero_decimal)
print("El número convertido en cadena es:", nueva_cadena)

# int() a entero
# float() a decimal
# str() a cadena
```
